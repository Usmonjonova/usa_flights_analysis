```python
def analyze_columns(df, missing_threshold=0.9, unique_threshold=1):
    
    # o'chirish uchun keraksiz columnlarni aniqlanadi
    
    report = []
    
    for col in df.columns:
        missing_pct = df[col].isnull().sum() / len(df)
        unique_count = df[col].nunique()
        unique_pct = unique_count / len(df)
        
        if missing_pct > missing_threshold:
            status = f"High missing ({missing_pct:.1%})"
        elif unique_count <= unique_threshold:
            status = "Zero variance"
        elif unique_pct > 0.99:
            status = "Likely ID column"
        else:
            status = "Keep"
        
        report.append({
            'Column': col,
            'Type': str(df[col].dtype),
            'Missing%': f"{missing_pct:.2%}",
            'Unique': unique_count,
            'Status': status
        })
    
    return pd.DataFrame(report)

column_report = analyze_columns(flights)
print(column_report.to_string())

cols_to_drop = column_report[column_report['Status'] != 'Keep']['Column'].tolist()
print(f"\nColumns to drop: {cols_to_drop}")
```

```python
# datasetdagi hamma year valuelari bir hil bo'lgani uchun uni o'chirib yuborish

flights['YEAR'].nunique()

flights = flights.drop(columns=['YEAR'])
```

```python
#  data turi son bolgan columnlardagi bo'sh cellarni 0 bilan to'ldirish

delay_columns = [
    'DEPARTURE_DELAY',
    'ARRIVAL_DELAY',
    'AIR_SYSTEM_DELAY',
    'SECURITY_DELAY',
    'AIRLINE_DELAY',
    'LATE_AIRCRAFT_DELAY',
    'WEATHER_DELAY'
]

for col in delay_columns:
    if col in flights.columns:
        flights[col] = flights[col].fillna(0)
```

```python
# CANCELLATION_REASON NaN bo'lganda uni Not Cancelled bilan almashtirish

flights['CANCELLATION_REASON'] = flights['CANCELLATION_REASON'].fillna('Not Cancelled')
```

```python
# TAIL_NUMBER da 14721 ta NaN borligi uchun shu rowlar o'chirib tashlandi

flights = flights.dropna(subset=['TAIL_NUMBER'])
```

```python
# 19. Oylar bo‘yicha o‘rtacha kechikishni line graph orqali tasvirlang (title, axis label va grid bilan). 


avg_delay_per_month = flights.groupby('MONTH_NAME')['DEPARTURE_DELAY'].mean().reset_index()

plt.figure(figsize=(12,6))
plt.plot(avg_delay_per_month['MONTH_NAME'], avg_delay_per_month['DEPARTURE_DELAY'], marker='o', linestyle='-', color='blue')
plt.title('Oylar boyicha ortacha chiqish kechikishi', fontsize=14)
plt.xlabel('Oy', fontsize=12)
plt.ylabel('DEPARTURE_DELAY', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 20.Eng yaxshi 5 ta aviakompaniyani (eng kam kechikish) bar chart orqali ko‘rsating. 

best_airlines = flights.groupby('AIRLINE_y')['DEPARTURE_DELAY'].min().reset_index().head(5)


plt.figure(figsize=(12,6))
plt.bar(best_airlines['AIRLINE_y'],
        best_airlines['DEPARTURE_DELAY'])

plt.title('Eng yaxshi 5 ta aviakompaniyani (eng kam kechikish)')
plt.xlabel('AIRLINE')
plt.ylabel('DEPARTURE_DELAY')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
# -80 eng yaxshi airline
```
